```python
for region in ['cleveland','switzerland','hungarian']:
    input_file = os.path.join(base_path, f'processed.{region}.data')
    output_files = os.path.join(base_path, f'{region}_heart_disease.csv')

    try:
        df = pd.read_csv(input_file, names=columns, na_values='?')
        df.to_csv(output_files, index = False)
        print(f'{region} 데이터 변환 완료 : {output_files}')
    except Exception as e:
        print(f"{region} 오류:",e)
```

```python
# 컬럼명 준비 (SPECT 데이터는 22개의 feature + 1개의 label)
columns = ['target'] + [f'F{i}' for i in range(1, 23)]

# 훈련 데이터와 테스트 데이터 읽기
try:
    train_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/SPECT.train', 
                          names=columns)
    test_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/SPECT.test', 
                          names=columns)
    
    # 전체 데이터 합치기 (나중에 필요하다면)
    spect_full = pd.concat([train_df, test_df], axis=0, ignore_index=True)
    
    print("Train 데이터 shape:", train_df.shape)
    print("Test 데이터 shape:", test_df.shape)
    print("전체 데이터 shape:", spect_full.shape)
    print("\n처음 5행:")
    print(spect_full.head())
    print("\n레이블 분포:")
    print(spect_full['target'].value_counts())
    
except Exception as e:
    print("파일 읽기 중 오류 발생:", e)
```

```python
# 기존 데이터 다시 읽기
cleveland_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/cleveland_heart_disease.csv')
hungarian_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/hungarian_heart_disease.csv')
swiss_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/switzerland_heart_disease.csv')

# Binary로 변환 (0은 그대로 0, 1-4는 1로 변환)
def to_binary(x):
    return 1 if x > 0 else 0

# 각 데이터셋에 대해 새로운 binary 레이블 컬럼 추가
cleveland_df['target_binary'] = cleveland_df['num'].apply(to_binary)
hungarian_df['target_binary'] = hungarian_df['num'].apply(to_binary)
swiss_df['target_binary'] = swiss_df['num'].apply(to_binary)

# 결과 확인
print("Cleveland binary label 분포:")
print(cleveland_df['target_binary'].value_counts())
print("\nHungarian binary label 분포:")
print(hungarian_df['target_binary'].value_counts())
print("\nSwiss binary label 분포:")
print(swiss_df['target_binary'].value_counts())

# 새로운 CSV 파일로 저장
cleveland_df.to_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/cleveland_binary.csv', index=False)
hungarian_df.to_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/hungarian_binary.csv', index=False)
swiss_df.to_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/heart_disease/switzerland_binary.csv', index=False)
```
