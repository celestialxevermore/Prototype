```python
# 데이터 읽기
heart_df = pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/heart.csv')

# HeartDisease 컬럼을 target_binary로 변환
heart_df['target_binary'] = heart_df['HeartDisease'].map({'no': 0, 'yes': 1})

# HeartDisease 컬럼과 Unnamed 컬럼들 삭제
heart_df = heart_df.drop('HeartDisease', axis=1)
# Unnamed로 시작하는 모든 컬럼 삭제
heart_df = heart_df.loc[:, ~heart_df.columns.str.contains('^Unnamed')]

# CSV로 저장
heart_df.to_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/heart_binary.csv', index=False)

# 확인
print("데이터 컬럼들:", heart_df.columns.tolist())
print("\nBinary label 분포:")
print(heart_df['target_binary'].value_counts())
```
