```python
import pandas as pd
import numpy as np

# 데이터셋 리스트
datasets = {
    'Cleveland': pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/cleveland_binary.csv'),
    'Hungarian': pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/hungarian_binary.csv'),
    'Switzerland': pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/switzerland_binary.csv'),
    'Statlog': pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/heart_statlog.csv'),
    'Heart': pd.read_csv('/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts/heart_binary.csv')
}

for name, df in datasets.items():
    print(f"\n{name} 데이터셋")
    print("="*50)
    print(f"데이터셋 크기: {df.shape}")
    
    # 결측치 통계
    missing_stats = pd.DataFrame({
        '전체 개수': len(df),
        '결측치 개수': df.isnull().sum(),
        '결측치 비율(%)': (df.isnull().sum() / len(df) * 100).round(2)
    })
    
    print("\n결측치 통계:")
    print(missing_stats[missing_stats['결측치 개수'] > 0])  # 결측치가 있는 컬럼만 표시
    
    # 데이터 타입 정보
    print("\n데이터 타입 정보:")
    print(df.dtypes)
    print("\n" + "="*50)
```

```python
def process_all_datasets(datasets):
    # 결측치가 많은 컬럼들 (Hungarian, Switzerland 기준)
    high_missing_cols = ['slope', 'ca', 'thal']
    
    # 각 데이터셋 복사
    processed_datasets = {}
    for name, df in datasets.items():
        processed_df = df.copy()
        
        # Cleveland, Hungarian, Switzerland 처리
        if name in ['Cleveland', 'Hungarian', 'Switzerland']:
            # 결측치가 많은 컬럼 제거
            processed_df = processed_df.drop(columns=high_missing_cols)
            
            # 남은 결측치 처리
            numeric_cols = ['trestbps', 'chol', 'thalach', 'oldpeak']
            categorical_cols = ['fbs', 'restecg', 'exang']
            
            # 수치형 컬럼 평균값으로 대체
            for col in numeric_cols:
                if col in processed_df.columns:
                    processed_df[col].fillna(processed_df[col].mean(), inplace=True)
            
            # 범주형 컬럼 최빈값으로 대체
            for col in categorical_cols:
                if col in processed_df.columns:
                    processed_df[col].fillna(processed_df[col].mode()[0], inplace=True)
        
        # Statlog 데이터셋 처리 (이미 결측치 없음)
        elif name == 'Statlog':
            pass
        
        # Heart 데이터셋 처리 (이미 결측치 없음)
        elif name == 'Heart':
            pass
            
        processed_datasets[name] = processed_df
    
    return processed_datasets

# 데이터셋 처리
processed_datasets = process_all_datasets(datasets)

# 처리 결과 확인
for name, df in processed_datasets.items():
    print(f"\n{name} 데이터셋 처리 결과")
    print("="*50)
    print(f"처리된 데이터셋 크기: {df.shape}")
    
    # 결측치 확인
    missing_stats = pd.DataFrame({
        '전체 개수': len(df),
        '결측치 개수': df.isnull().sum(),
        '결측치 비율(%)': (df.isnull().sum() / len(df) * 100).round(2)
    })
    
    print("\n결측치 통계:")
    print(missing_stats[missing_stats['결측치 개수'] > 0])  # 결측치가 있는 컬럼만 표시
    print("\n" + "="*50)
```

```python
import os

# 저장 경로 설정
save_dir = '/home/eungyeop/LLM/tabular/ProtoLLM/dataset/data/table/hearts_preprocessed'

# 디렉토리가 없으면 생성
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

# 각 처리된 데이터셋 저장
for name, df in processed_datasets.items():
    # 파일명 설정
    if name == 'Cleveland':
        filename = 'cleveland_binary_preprocessed.csv'
    elif name == 'Hungarian':
        filename = 'hungarian_binary_preprocessed.csv'
    elif name == 'Switzerland':
        filename = 'switzerland_binary_preprocessed.csv'
    elif name == 'Statlog':
        filename = 'heart_statlog_preprocessed.csv'
    else:  # Heart
        filename = 'heart_binary_preprocessed.csv'
    
    # 전체 파일 경로
    file_path = os.path.join(save_dir, filename)
    
    # CSV로 저장
    df.to_csv(file_path, index=False)
    print(f"{name} 데이터셋이 {file_path}에 저장되었습니다.")
```
