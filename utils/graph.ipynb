```python
import matplotlib.pyplot as plt
import torch
import numpy as np

def visualize_graph_structure(data, title):
    plt.figure(figsize=(10, 10))
    plt.title(title)
    
    # 노드 수
    n_nodes = data.x.shape[0]
    
    # 중심 노드를 가운데에 두고, 나머지 노드들을 원형으로 배치
    angles = np.linspace(0, 2*np.pi, n_nodes-1, endpoint=False)
    
    # 노드 위치 계산
    pos = np.zeros((n_nodes, 2))
    pos[0] = [0, 0]  # 중심 노드
    radius = 1
    for i in range(1, n_nodes):
        pos[i] = [radius * np.cos(angles[i-1]), radius * np.sin(angles[i-1])]
    
    # edge 그리기
    for i in range(data.edge_index.shape[1]):
        start = data.edge_index[0, i]
        end = data.edge_index[1, i]
        plt.plot([pos[start, 0], pos[end, 0]], 
                [pos[start, 1], pos[end, 1]], 
                'gray', alpha=0.5)
    
    # 노드 그리기
    plt.scatter(pos[0, 0], pos[0, 1], c='red', s=200, label='Center Node')  # 중심 노드
    plt.scatter(pos[1:, 0], pos[1:, 1], c='blue', s=200, label='Feature Nodes')  # feature 노드
    
    # 노드 레이블 추가
    features = ['Age', 'BP', 'Height', 'Weight', 'Glucose', 'Gender']
    plt.annotate('Center', (pos[0, 0], pos[0, 1]))
    for i in range(1, n_nodes):
        plt.annotate(features[i-1], (pos[i, 0], pos[i, 1]))
    
    plt.legend()
    plt.axis('equal')
    plt.show()

# Basic version과 Leaf-to-leaf version 시각화
visualize_graph_structure(graph_data[0], "Basic Graph Structure (Star Graph)")
visualize_graph_structure(graph_l2l[0], "Graph with Leaf-to-Leaf Connections")
```

```python
def visualize_attention_graph(data):
    plt.figure(figsize=(12, 12))
    plt.title("Graph with Attention Weights")
    
    n_nodes = data.x.shape[0]
    angles = np.linspace(0, 2*np.pi, n_nodes-1, endpoint=False)
    
    # 노드 위치 계산
    pos = np.zeros((n_nodes, 2))
    pos[0] = [0, 0]
    radius = 1
    for i in range(1, n_nodes):
        pos[i] = [radius * np.cos(angles[i-1]), radius * np.sin(angles[i-1])]
    
    # Edge와 attention weight 시각화
    if hasattr(data, 'attention'):
        attention_weights = data.attention
        # attention이 매트릭스인 경우 처리
        if len(attention_weights.shape) > 1:
            attention_weights = attention_weights.mean(dim=1)
            
        vmin, vmax = attention_weights.min(), attention_weights.max()
        cmap = plt.cm.RdYlBu
        
        for i in range(data.edge_index.shape[1]):
            start = data.edge_index[0, i]
            end = data.edge_index[1, i]
            
            # attention weight 정규화
            weight = attention_weights[i].item()
            norm_weight = (weight - vmin) / (vmax - vmin)
            
            # Edge 그리기
            width = 1 + 3 * norm_weight
            color = cmap(norm_weight)
            
            plt.plot([pos[start, 0], pos[end, 0]], 
                    [pos[start, 1], pos[end, 1]], 
                    c=color, linewidth=width, alpha=0.7)
    
    # 노드 그리기
    plt.scatter(pos[0, 0], pos[0, 1], c='red', s=300, label='Center Node')
    plt.scatter(pos[1:, 0], pos[1:, 1], c='blue', s=200, label='Feature Nodes')
    
    # 노드 레이블
    features = ['Age', 'BP', 'Height', 'Weight', 'Glucose', 'Gender']
    plt.annotate('Center', (pos[0, 0], pos[0, 1]))
    for i in range(1, n_nodes):
        plt.annotate(features[i-1], (pos[i, 0], pos[i, 1]))
    
    plt.colorbar(plt.cm.ScalarMappable(cmap=cmap), label='Attention Weight')
    plt.legend()
    plt.axis('equal')

def visualize_message_graph(data):
    plt.figure(figsize=(12, 12))
    plt.title("Graph with Message Importance")
    
    n_nodes = data.x.shape[0]
    angles = np.linspace(0, 2*np.pi, n_nodes-1, endpoint=False)
    
    pos = np.zeros((n_nodes, 2))
    pos[0] = [0, 0]
    radius = 1
    for i in range(1, n_nodes):
        pos[i] = [radius * np.cos(angles[i-1]), radius * np.sin(angles[i-1])]
    
    if hasattr(data, 'H'):
        # Message importance 계산
        H = data.H
        message_importance = torch.sum(H, dim=1)
        norm_importance = (message_importance - message_importance.min()) / (
            message_importance.max() - message_importance.min())
        
        # Edge 그리기
        for i in range(data.edge_index.shape[1]):
            start = data.edge_index[0, i]
            end = data.edge_index[1, i]
            plt.plot([pos[start, 0], pos[end, 0]], 
                    [pos[start, 1], pos[end, 1]], 
                    'gray', alpha=0.3)
        
        # 노드 그리기
        sizes = 200 + 1000 * norm_importance
        scatter = plt.scatter(pos[:, 0], pos[:, 1], 
                            s=sizes, 
                            c=message_importance,
                            cmap='viridis', 
                            alpha=0.6)
        
        # 노드 레이블
        features = ['Age', 'BP', 'Height', 'Weight', 'Glucose', 'Gender']
        plt.annotate('Center', (pos[0, 0], pos[0, 1]))
        for i in range(1, n_nodes):
            plt.annotate(features[i-1], (pos[i, 0], pos[i, 1]))
        
        plt.colorbar(scatter, label='Message Importance')
    
    plt.axis('equal')

# 시각화 실행
visualize_attention_graph(graphs_attention[0])
plt.show()

visualize_message_graph(graphs_message[0])
plt.show()
```

```python
def visualize_attention_graph(data):
   plt.figure(figsize=(12, 12))
   plt.title("Graph with Attention Weights")
   
   n_nodes = data.x.shape[0]
   angles = np.linspace(0, 2*np.pi, n_nodes-1, endpoint=False)
   
   # 노드 위치 계산
   pos = np.zeros((n_nodes, 2))
   pos[0] = [0, 0]  # center node
   radius = 1
   for i in range(1, n_nodes):
       pos[i] = [radius * np.cos(angles[i-1]), radius * np.sin(angles[i-1])]
   
   # Edge와 attention weight 시각화
   if hasattr(data, 'attention'):
       attention_weights = data.attention
       if len(attention_weights.shape) > 1:
           attention_weights = attention_weights.mean(dim=1)
           
       vmin, vmax = attention_weights.min(), attention_weights.max()
       cmap = plt.cm.RdYlBu
       
       # Edge 먼저 그리기
       for i in range(data.edge_index.shape[1]):
           start = data.edge_index[0, i]
           end = data.edge_index[1, i]
           
           weight = attention_weights[i].item()
           norm_weight = (weight - vmin) / (vmax - vmin)
           
           # Edge 특성 강화
           width = 2 + 4 * norm_weight  # 선 두께 증가
           color = cmap(norm_weight)
           
           plt.plot([pos[start, 0], pos[end, 0]], 
                   [pos[start, 1], pos[end, 1]], 
                   c=color, linewidth=width, alpha=0.7, zorder=1)
   
   # 노드는 edge 위에 그리기
   plt.scatter(pos[0, 0], pos[0, 1], c='red', s=300, label='Center Node', zorder=2)
   plt.scatter(pos[1:, 0], pos[1:, 1], c='blue', s=200, label='Feature Nodes', zorder=2)
   
   features = ['Age', 'BP', 'Height', 'Weight', 'Glucose', 'Gender']
   plt.annotate('Center', (pos[0, 0], pos[0, 1]), zorder=3)
   for i in range(1, n_nodes):
       plt.annotate(features[i-1], (pos[i, 0], pos[i, 1]), zorder=3)
   
   plt.colorbar(plt.cm.ScalarMappable(cmap=cmap), label='Attention Weight')
   plt.legend()
   plt.axis('equal')

def visualize_message_graph(data):
   plt.figure(figsize=(12, 12))
   plt.title("Graph with Message Importance")
   
   n_nodes = data.x.shape[0]
   angles = np.linspace(0, 2*np.pi, n_nodes-1, endpoint=False)
   
   pos = np.zeros((n_nodes, 2))
   pos[0] = [0, 0]
   radius = 1
   for i in range(1, n_nodes):
       pos[i] = [radius * np.cos(angles[i-1]), radius * np.sin(angles[i-1])]
   
   if hasattr(data, 'H'):
       H = data.H
       message_importance = torch.sum(H, dim=1)
       norm_importance = (message_importance - message_importance.min()) / (
           message_importance.max() - message_importance.min())
       
       # Edge 먼저 그리기 - 더 선명하게
       for i in range(data.edge_index.shape[1]):
           start = data.edge_index[0, i]
           end = data.edge_index[1, i]
           
           # message importance에 따른 edge 두께 조절
           edge_importance = (norm_importance[start] + norm_importance[end]) / 2
           width = 1 + 3 * edge_importance
           
           plt.plot([pos[start, 0], pos[end, 0]], 
                   [pos[start, 1], pos[end, 1]], 
                   'gray', alpha=0.4, linewidth=width.item(), zorder=1)
       
       # 노드 크기 더 차별화
       sizes = 200 + 2000 * norm_importance
       scatter = plt.scatter(pos[:, 0], pos[:, 1], 
                           s=sizes, 
                           c=message_importance,
                           cmap='viridis', 
                           alpha=0.8,
                           zorder=2)
       
       features = ['Age', 'BP', 'Height', 'Weight', 'Glucose', 'Gender']
       plt.annotate('Center', (pos[0, 0], pos[0, 1]), zorder=3)
       for i in range(1, n_nodes):
           plt.annotate(features[i-1], (pos[i, 0], pos[i, 1]), zorder=3)
       
       plt.colorbar(scatter, label='Message Importance')
   
   plt.axis('equal')

# 시각화 실행
visualize_attention_graph(graphs_attention[0])
plt.show()

visualize_message_graph(graphs_message[0])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer

# 1. 원본 데이터 분포
numerical_columns = ['age', 'blood_pressure', 'height', 'weight', 'glucose']

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Distribution Comparison: Original vs StandardScaler vs PowerTransformer', y=1.05, size=15)

# Original Data
for idx, col in enumerate(numerical_columns):
    ax = axes[0, idx%2] if idx < 2 else axes[1, idx%2] if idx < 4 else axes[2, 0]
    sns.histplot(data=df, x=col, ax=ax, kde=True)
    ax.set_title(f'Original {col} Distribution')

# 2. StandardScaler 변환 (우리의 새 버전)
transformed_standard = pd.DataFrame()
for col in numerical_columns:
    transformed_standard[col] = transformer.scalers[col].transform(df[col].values.reshape(-1, 1)).flatten()

# StandardScaler 결과
for idx, col in enumerate(numerical_columns):
    ax = axes[0, idx%2] if idx < 2 else axes[1, idx%2] if idx < 4 else axes[2, 0]
    sns.histplot(data=transformed_standard, x=col, ax=ax, color='red', alpha=0.5, kde=True)
    ax.set_title(f'{col} Distribution (Blue: Original, Red: StandardScaler)')

# 3. PowerTransformer 변환 (원래 버전)
pt = PowerTransformer()
transformed_power = pd.DataFrame(
    pt.fit_transform(df[numerical_columns]), 
    columns=numerical_columns
)

# PowerTransformer 결과 추가
for idx, col in enumerate(numerical_columns):
    ax = axes[0, idx%2] if idx < 2 else axes[1, idx%2] if idx < 4 else axes[2, 0]
    sns.histplot(data=transformed_power, x=col, ax=ax, color='green', alpha=0.3, kde=True)
    ax.set_title(f'{col} Distribution\n(Blue: Original, Red: StandardScaler, Green: PowerTransformer)')

plt.tight_layout()
plt.show()

# 기술 통계량 비교
print("\nOriginal Data Statistics:")
print(df[numerical_columns].describe())

print("\nStandardScaler Transformed Data Statistics:")
print(transformed_standard.describe())

print("\nPowerTransformer Transformed Data Statistics:")
print(transformed_power.describe())
```
