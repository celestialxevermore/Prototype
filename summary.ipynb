```python
import json
import csv
import glob
import os
import numpy as np
from collections import defaultdict

def process_json_files(directory_path):
    json_files = glob.glob(os.path.join("/home", "eungyeop", directory_path, "**/*.json"), recursive=True)
    
    if not json_files:
        print("JSON 파일을 찾을 수 없습니다.")
        return
    
    # 각 JSON 파일별로 개별 처리
    for json_file in json_files:
        with open(json_file, 'r') as f:
            data = json.load(f)
            
            # CSV 파일 경로 설정
            dir_path = os.path.dirname(json_file)
            base_name = os.path.splitext(os.path.basename(json_file))[0]
            output_file = os.path.join(dir_path, f"{base_name}.csv")
            
            with open(output_file, 'w', newline='') as f:
                writer = csv.writer(f)
                
                # 헤더 작성
                writer.writerow(['Model Type', data['model_type']])
                writer.writerow(['Graph Type', data['graph_type']])
                writer.writerow(['Center Type', data['center_type']])
                writer.writerow(['Few Shot', data['hyperparameters']['few_shot']])
                writer.writerow(['Batch Size', data['hyperparameters']['batch_size']])
                writer.writerow(['Num Layers', data['hyperparameters']['num_layers']])
                writer.writerow(['Heads', data['hyperparameters']['heads']])
                writer.writerow([''])  # 빈 줄 추가
                
                # 해당 JSON의 결과값 작성
                if 'Ours_few' in data['results']:
                    results = data['results']['Ours_few']
                    writer.writerow([f"{results['Ours_best_few_auc']:.4f}"])
                    writer.writerow([f"{results['Ours_best_few_acc']:.4f}"])
                    writer.writerow([f"{results['Ours_best_few_precision']:.4f}"])
                    writer.writerow([f"{results['Ours_best_few_recall']:.4f}"])
                    writer.writerow([f"{results['Ours_best_few_f1']:.4f}"])
            
            print(f"결과가 {output_file}에 저장되었습니다.")

# 사용 예시
directory_path = "LLM/tabular/ProtoLLM/experiments/source_to_source_SEEDS/heart"
process_json_files(directory_path)
```

```python
import json
import csv
import glob
import os
import numpy as np
from collections import defaultdict

def create_directory(path):
    if not os.path.exists(path):
        os.makedirs(path)

def process_json_files(directory_path):
    # 실험 설정별로 결과를 모을 딕셔너리
    results_by_config = defaultdict(lambda: defaultdict(list))
    
    # JSON 파일 찾기 (전체 경로 사용)
    json_pattern = os.path.join("/home", "eungyeop", directory_path, "**/GAT_edge*/star_*/f*.json")
    json_files = glob.glob(json_pattern, recursive=True)
    
    if not json_files:
        print("JSON 파일을 찾을 수 없습니다.")
        return
    
    # 먼저 모든 결과를 수집
    for json_file in json_files:
        try:
            with open(json_file, 'r') as f:
                data = json.load(f)
                
                # 파일 경로에서 필요한 정보 추출
                path_parts = json_file.split('/')
                model_type = path_parts[-3]  # GAT_edge
                graph_center_type = path_parts[-2]  # star_D_CM
                
                # 실험 설정을 키로 사용
                config_key = (
                    model_type,
                    graph_center_type,
                    data['hyperparameters']['few_shot'],
                    data['hyperparameters']['batch_size'],
                    data['hyperparameters']['num_layers'],
                    data['hyperparameters']['heads']
                )
                
                if 'Ours_few' in data['results']:
                    results = data['results']['Ours_few']
                    results_by_config[config_key]['auc'].append(results['Ours_best_few_auc'])
                    results_by_config[config_key]['acc'].append(results['Ours_best_few_acc'])
                    results_by_config[config_key]['precision'].append(results['Ours_best_few_precision'])
                    results_by_config[config_key]['recall'].append(results['Ours_best_few_recall'])
                    results_by_config[config_key]['f1'].append(results['Ours_best_few_f1'])
                    
                    # 디버깅을 위한 출력
                    print(f"\nProcessing file: {json_file}")
                    print(f"AUC value: {results['Ours_best_few_auc']}")
        except Exception as e:
            print(f"Error processing file {json_file}: {str(e)}")
    
    # 결과를 새로운 디렉토리 구조로 저장
    base_dir = os.path.join("/home", "eungyeop", directory_path)
    summary_dir = os.path.join(base_dir, "summary")
    
    for config_key, results in results_by_config.items():
        try:
            model_type, graph_center_type, few_shot, batch_size, num_layers, heads = config_key
            
            # 디렉토리 구조 생성
            model_dir = os.path.join(summary_dir, model_type)
            graph_dir = os.path.join(model_dir, graph_center_type)
            create_directory(graph_dir)
            
            # CSV 파일 이름 생성
            filename = f"f{few_shot}_b{batch_size}_l{num_layers}_h{heads}.csv"
            output_file = os.path.join(graph_dir, filename)
            
            with open(output_file, 'w', newline='') as f:
                writer = csv.writer(f)
                
                # 헤더 작성
                writer.writerow(['Model Type', model_type])
                writer.writerow(['Graph Type', graph_center_type.split('_')[0]])
                writer.writerow(['FD Type', graph_center_type.split('_')[1]])
                writer.writerow(['Center Type', '_'.join(graph_center_type.split('_')[2:])])
                writer.writerow(['Few Shot', few_shot])
                writer.writerow(['Batch Size', batch_size])
                writer.writerow(['Num Layers', num_layers])
                writer.writerow(['Heads', heads])
                writer.writerow([''])
                
                # 디버깅을 위한 출력
                print(f"\nGenerating summary for config: {config_key}")
                print("Current values for each metric:")
                
                # 평균과 표준편차 계산하여 작성
                metrics = ['auc', 'acc', 'precision', 'recall', 'f1']
                for metric in metrics:
                    values = np.array(results[metric])
                    print(f"{metric}: {values}")
                    mean = np.mean(values)
                    std = np.std(values)
                    writer.writerow([f"{mean:.4f}({std:.4f})"])
                    print(f"{metric} summary: {mean:.4f}({std:.4f})")
            
            print(f"결과가 {output_file}에 저장되었습니다.")
        except Exception as e:
            print(f"Error processing config {config_key}: {str(e)}")

# 사용 예시
directory_path = "LLM/tabular/ProtoLLM/experiments/source_to_source_SEEDS/adult"
process_json_files(directory_path)
```

```python
import json
import csv
import glob
import os
import numpy as np
from collections import defaultdict

def create_directory(path):
    if not os.path.exists(path):
        os.makedirs(path)

def calculate_mean_std(values):
    """평균과 표준편차를 계산"""
    values = np.array(values)
    mean = np.mean(values)
    std = np.std(values, ddof=0)  # ddof=0으로 설정하여 모집단 표준편차 계산
    return mean, std

def process_json_files(directory_path):
    results_by_config = defaultdict(lambda: defaultdict(list))
    
    json_pattern = os.path.join("/home", "eungyeop", directory_path, "**/GAT_edge*/star_*/f*.json")
    json_files = glob.glob(json_pattern, recursive=True)
    
    if not json_files:
        print("JSON 파일을 찾을 수 없습니다.")
        return
    
    # 결과 수집
    for json_file in json_files:
        with open(json_file, 'r') as f:
            data = json.load(f)
            
            path_parts = json_file.split('/')
            model_type = path_parts[-3]
            graph_center_type = path_parts[-2]
            
            config_key = (
                model_type,
                graph_center_type,
                data['hyperparameters']['few_shot'],
                data['hyperparameters']['batch_size'],
                data['hyperparameters']['num_layers'],
                data['hyperparameters']['heads']
            )
            
            if 'Ours_few' in data['results']:
                results = data['results']['Ours_few']
                results_by_config[config_key]['auc'].append(results['Ours_best_few_auc'])
                results_by_config[config_key]['acc'].append(results['Ours_best_few_acc'])
                results_by_config[config_key]['precision'].append(results['Ours_best_few_precision'])
                results_by_config[config_key]['recall'].append(results['Ours_best_few_recall'])
                results_by_config[config_key]['f1'].append(results['Ours_best_few_f1'])
                
                print(f"\n처리 중인 파일: {json_file}")
                print(f"AUC: {results['Ours_best_few_auc']}")
    
    # 결과 저장
    base_dir = os.path.join("/home", "eungyeop", directory_path)
    summary_dir = os.path.join(base_dir, "summary")
    
    for config_key, results in results_by_config.items():
        model_type, graph_center_type, few_shot, batch_size, num_layers, heads = config_key
        
        model_dir = os.path.join(summary_dir, model_type)
        graph_dir = os.path.join(model_dir, graph_center_type)
        create_directory(graph_dir)
        
        filename = f"f{few_shot}_b{batch_size}_l{num_layers}_h{heads}.csv"
        output_file = os.path.join(graph_dir, filename)
        
        with open(output_file, 'w', newline='') as f:
            writer = csv.writer(f)
            
            writer.writerow(['Model Type', model_type])
            writer.writerow(['Graph Type', graph_center_type.split('_')[0]])
            writer.writerow(['FD Type', graph_center_type.split('_')[1]])
            writer.writerow(['Center Type', '_'.join(graph_center_type.split('_')[2:])])
            writer.writerow(['Few Shot', few_shot])
            writer.writerow(['Batch Size', batch_size])
            writer.writerow(['Num Layers', num_layers])
            writer.writerow(['Heads', heads])
            writer.writerow([''])
            
            print(f"\n설정 {config_key}에 대한 요약 생성 중")
            
            metrics = ['auc', 'acc', 'precision', 'recall', 'f1']
            for metric in metrics:
                values = results[metric]
                print(f"{metric} 값들: {values}")
                mean, std = calculate_mean_std(values)
                writer.writerow([f"{mean:.4f}({std:.4f})"])
                print(f"{metric} 요약: {mean:.4f}({std:.4f})")
        
        print(f"결과가 {output_file}에 저장되었습니다.")

# 사용 예시
directory_path = "LLM/tabular/ProtoLLM/experiments/source_to_source_SEEDS/diabetes"
process_json_files(directory_path)
```
